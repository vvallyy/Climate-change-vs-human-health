```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("heat-deaths_fig-2.csv", encoding = 'cp1252')
```

```python
df
```

```python
df1 = df.drop(index=[0,1,2,3,4])
```

```python
df1
```

```python
df1.columns = df1.loc[5]
```

```python
df1
```

```python
df1 = df1.drop(index = [5])
```

```python
df1
```

```python
df1 = df1.reset_index()
```

```python
df1
```

```python
df1 = df1.drop(['index'], axis = 1)
```

```python
df1
```

```python
df2 = df1.iloc[:,1].astype('float64')
df3 = df1.iloc[:,2].astype('float64')
df4 = df1.iloc[:,3].astype('float64')
df5 = df1.iloc[:,4].astype('float64')
```

```python
plt.plot(df1['Year'], df2, label = 	'Crude summer death rate per million, general population')
plt.plot(df1['Year'], df3, label = 'Crude summer death rate per million, age 65+ population')

plt.plot(df1['Year'], df4, label = 'Crude summer death rate per million, non-white population')
plt.plot(df1['Year'], df5, label = 'Crude summer death rate per million, non-Hispanic black population')
plt.grid(alpha = 0.5)
plt.title("Summer Deaths Due to Heat and Cardiovascular Disease in the United States, 1999â€“2018")
plt.xlabel("Year")
plt.ylabel('Death Rate (per million people)', fontsize = 'x-small')
legend = plt.legend(loc = 'upper right', shadow=True, fontsize = 'x-small')
ax = plt.gca()
for tick in ax.get_xticklabels():
    tick.set_rotation(45)
plt.show()
#plt.plot(df1['Year'], df1['Underlying and contributing causes of death (May-Sept)'])
```
